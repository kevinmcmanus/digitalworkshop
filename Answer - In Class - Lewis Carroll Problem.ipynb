# The Lewis Carroll Problem

Lewis Carroll had a bunch of problems that kept him up at night. Here's one that I'd like you to simulate.

You have a bag.

The bag has 1 ball in it. 

That ball is either red or green (50/50 chance).

I add a ***red*** ball to the bag.

I randomly select a ball out of the bag.

I show you the ball I pulled out is ***red***.

What is the probability that the ball inside the bag is ***red***?


# Solution

```
import random

# "red_pulled" and "green_pulled" represent the second ball pulled (the one we care about being red or green)
# We will start at 0 and then add 1 to them each time they are pulled from the bag
red_pulled = 0
green_pulled = 0

for ball in range(10000):
  # Create a bag with 1 red ball, then a 50/50 chance of red/green
  bag = ['red', random.choice(['red', 'green'])]

  # Shuffle the balls in the bag
  random.shuffle(bag)

  # Check to see if the first ball pulled is red
  if bag[0] == 'red':
    if bag[1] == 'red':
      red_pulled += 1
    else:
      green_pulled += 1
  else:
    pass # This condition will only be satisfied if the first ball pulled is green, in that case, we don't care

red_probability = red_pulled / (red_pulled + green_pulled)

print(f"Probability of the 2nd ball being red is {round(100*red_probability, 2)}%")
```
